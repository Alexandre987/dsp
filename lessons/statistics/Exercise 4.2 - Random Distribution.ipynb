**Exercise:** The numbers generated by `numpy.random.random` are supposed to be uniform between 0 and 1; that is, every value in the range should have the same probability.

Generate 1000 numbers from `numpy.random.random` and plot their PMF.  What goes wrong?

Now plot the CDF. Is the distribution uniform?

```python
nums = np.random.random(1000)
```

```python
pmf = thinkstats2.Pmf(nums)
```

```python
pmf
```

In the above case, every individual random number has its own frequency, and it's likely that no number is repeated (or if they are repeated, it is sporadic). But this says nothing about the overall distribution of the data, and whether the values will be more concentrated in any region or follow some sort of pattern. I need to use the CDF so that the intense granularity of the data doesn't cause a problem

```python
cdf = thinkstats2.Cdf(nums,label='cumulative_numbers')
thinkplot.Cdf(cdf)
```

In the above distribution, we see that the CDF is linear, showing that the np.random.random function does seem to produce numbers randomly
